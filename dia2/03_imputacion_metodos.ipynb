```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import missingno
```

```python
diabetes_df = pd.read_csv('/content/diabetes.csv')
diabetes_df
```

```python
diabetes_df[diabetes_df.columns[1:6]] = diabetes_df[diabetes_df.columns[1:6]].replace(0,np.nan)
diabetes_df.isna().sum().sum()
```

## IMPUTACION POR MEDIA,MEDIANA,MODA

```python
diabetes_df.Glucose.isna().sum()
```

```python
sorted_df = diabetes_df.sort_values(
    by=['Glucose'],
    ascending=False
)
sorted_df
```

```python
sorted_df['Glucose'].mean()
```

```python
diabates_glucose_df = diabetes_df
diabates_glucose_df
```

```python
diabates_glucose_df["Glucose"] = diabates_glucose_df["Glucose"].fillna(diabates_glucose_df["Glucose"].mean())
sorted_glucose_df = diabates_glucose_df.sort_values(
    by=['Glucose'],
    ascending=False
)
sorted_glucose_df
```

```python
sorted_glucose_df['Glucose'].mean()
```

IMPUTACION POR MEDIANA

```python
diabetes_df['Glucose'].median()
```

```python
diabates_glucose_median_df = diabetes_df
diabates_glucose_median_df["Glucose"] = diabates_glucose_median_df["Glucose"].fillna(diabates_glucose_median_df["Glucose"].median())
sorted_glucose_median_df = diabates_glucose_median_df.sort_values(
    by=['Glucose'],
    ascending=False
)
sorted_glucose_median_df
```

```python
sorted_glucose_median_df['Glucose'].median()
```

## IMPUTACION POR BFILL Y FFILL ( LLENADO HACIA ADELANTE Y LLENADO HACIA ATRAS)

```python
sorted_skin_thickness_df = diabetes_df.sort_values(
    by=['Age','Glucose'],
    ascending=True
)
sorted_skin_thickness_df.SkinThickness.isna().sum()
```

```python
 sorted_skin_thickness_df.SkinThickness
```

```python
 sorted_skin_thickness_df.SkinThickness.ffill()
```
