1 - IMPORTAR LIBRERIAS

```python
import requests
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

2 - EXTRAER DATA DE RANDOMUSER

```python
url = 'https://randomuser.me/api/?results=100'
response = requests.get(url)
data = response.json()
print(data)
```

3 - CONVERTIR EL RESPONSE A DATAFRAME

```python
results = data['results']
lista_sexos = [user['gender'] for user in results]
print(lista_sexos)
lista_edades = [user['dob']['age'] for user in results]
print(lista_edades)
lista_paises = [user['location']['country'] for user in results]
```

```python
df_users = pd.DataFrame({
    'sexo':lista_sexos,
    'edad':lista_edades,
    'pais':lista_paises
})
df_users
```

CREAMOS GRAFICOS CON SEABORN

```python
sns.histplot(data=df_users,x='edad',hue='sexo',stat='count',multiple='dodge')
plt.show()
```

```python
sns.kdeplot(data=df_users,x='edad',hue='sexo',fill=True)
plt.show()
```

```python
sns.scatterplot(data=df_users,x=df_users.index,y='edad',hue='sexo')
plt.xlabel('Index')
plt.show()
```

```python
sns.pairplot(data=df_users,hue='sexo')
plt.show()
```
