```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy
import statsmodels.api as sm
import statsmodels.formula.api as smf
import statsmodels.stats as ss
```

```python
%matplotlib inline
sns.set_style(style='whitegrid')
sns.set_context(context='notebook')
plt.rcParams['figure.figsize'] = (11, 9.4)

penguin_color = {
    'Adelie': '#ff6602ff',
    'Gentoo': '#0f7175ff',
    'Chinstrap': '#c65dc9ff'
}
```

```python
penguins_df = sns.load_dataset('penguins')
penguins_df.dropna(inplace=True)
penguins_df
```

```python
sns.heatmap(
    data=penguins_df.corr(numeric_only=True),
    cmap=sns.diverging_palette(20, 230, as_cmap=True),
    center=0,
    vmin=-1,
    vmax=1,
    linewidths=0.5,
    cbar_kws={"shrink": 0.5},
    annot=True
)
```

```python
sns.scatterplot(
    data=penguins_df,
    x='bill_length_mm',
    y='bill_depth_mm'
)
```

```python
sns.regplot(
    data=penguins_df,
    x='bill_length_mm',
    y='bill_depth_mm',
)
```

```python
sns.lmplot(
    data=penguins_df,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='species',
    height=10,
    palette=penguin_color
)
```

```python
sns.pairplot(data=penguins_df,hue='species',palette=penguin_color)
```
