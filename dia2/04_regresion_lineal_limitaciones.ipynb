LIMITACIONES DE LA REGRESION LINEAL SIMPLE


1.   LA REGRESIÓN LINEAL DE X -Y NO ES IGUAL A Y-X
2.   SI DOS VARIABLES CRECEN O DECRECEN SIGUIENTO LAS MISMAS PAUTAS, NO IMPLICA CAUSALIDAD
3. SOLO SE PUEDE MANEJAR RELACIONES LINEALES

COMPROBAMOS QUE LA REGRESION LINEAL SIMPLE NO ES SIMETRICA

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy
```

```python
penguins_df = sns.load_dataset('penguins')
penguins_df.dropna(inplace=True)
penguins_df
```

REGRESIÓN LINEAL DE x = bill_length_mm | y = bill_depth_mm

```python
x = penguins_df.bill_length_mm
y = penguins_df.bill_depth_mm

res_x_y = scipy.stats.linregress(x=x,y=y)
res_y_x = scipy.stats.linregress(y=x,x=y)

print(res_x_y)
print(res_y_x)
```

GRAFICAMOS LAS 2 REGRESIONES LINEALES SIMPLES

```python
sns.scatterplot(
    x=x,
    y=y
)

fx_1 = np.array([x.min(),x.max()])
fy_1 = res_x_y.slope * fx_1 + res_x_y.intercept
plt.plot(fx_1,fy_1,color='red')
```

```python
sns.scatterplot(
    x=y,
    y=x
)
fx_2 = np.array([y.min(),y.max()])
fy_2 = res_y_x.slope * fx_2 + res_y_x.intercept

plt.plot(fx_2,fy_2,color='green')
```

```python
sns.scatterplot(
    x=x,
    y=y
)

plt.plot(fx_1,fy_1,color='red')
plt.plot(fy_2,fx_2,color='green')
```
