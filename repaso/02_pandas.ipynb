## TRABAJANDO CON VALORES NULOS

```python
df_penguins.isnull()
```

```python
df_penguins.bill_length_mm.isnull()
```

```python
mean_bill_length_mm = df_penguins.bill_length_mm.mean()
mean_bill_length_mm
```

```python
df_penguins.bill_length_mm.fillna(mean_bill_length_mm,inplace=True)
```

```python
df_penguins.bill_length_mm.isnull()
```

```python
df_penguins.dropna()
```
