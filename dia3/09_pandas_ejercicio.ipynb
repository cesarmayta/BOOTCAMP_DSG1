IMPORTAMOS LIBRERIAS

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

HACER LLAMADA A API Y OBTENER DATOS

```python
url = 'https://randomuser.me/api/?results=10'
response = requests.get(url)

if response.status_code == 200:
  data = response.json()
else:
  print(f"error al hacer request : {response.status_code}")
```

```python
users = data['results']
users
```

```python
df_users = pd.json_normalize(users)
df_users
```

```python
df_users.info()
```

FILTRAR SOLO SEXO Y EDAD

```python
df_sexo_edad = df_users[['gender','dob.age']]
df_sexo_edad.columns = ['Sexo','Edad']
df_sexo_edad
```

CALCULAR EDAD PROMEDIO POR SEXO

```python
edad_promedio_por_sexo = df_sexo_edad.groupby('Sexo')['Edad'].mean()
edad_promedio_por_sexo
```

CREARMOS GRAFICO

```python
edad_promedio_por_sexo.plot(kind='bar',color=['blue','pink'])
plt.title('Edad Promedio por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Edad Proemdio')
plt.show()
```
