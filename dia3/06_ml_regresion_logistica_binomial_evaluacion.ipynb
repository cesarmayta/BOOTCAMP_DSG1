# ELIMINAMOS EL ID

```python
df_data.drop('customerID',axis=1,inplace=True)
```

```python
df_data
```

# CODIFICACIÓN DE VARIABLES CATEGORICAS

```python
df_data_processing = df_data.copy()
```

```python
df_data_processing
```

CONVERTIR VARIABLE DE SALIDA CHURN EN NUMERICA

```python
df_data_processing['Churn'].replace(to_replace='Yes',value=1,inplace=True)
df_data_processing['Churn'].replace(to_replace='No',value=0,inplace=True)
df_data_processing.info()
```

## CONVERTIMOS VARIABLE TOTALCHARGE DE OBJECT A NUMERICA

```python
df_data_processing.TotalCharges = pd.to_numeric(df_data_processing.TotalCharges,errors='coerce')
```

```python
df_data_processing.info()
```

## PASO 1 - IDENTIFICAMOS VARIABLES CATEGORICAS

```python
categorical_columns = df_data_processing.select_dtypes(object).columns
categorical_columns
```

## PASO 2 - CREAMOS UN TRANSFORMADOR DE VARIABLES CATEGORICAS A NUMERICAS

```python
import sklearn.compose
import sklearn.preprocessing
```

```python
categorical_transformer = sklearn.compose.make_column_transformer(
    (sklearn.preprocessing.OneHotEncoder(), categorical_columns),
    remainder="passthrough"
)
categorical_transformer
```

## PASO 3 : CODIFICAMOS EL DATASET CON EL TRANSFORMADOR QUE CREAMOS

```python
df_data_processing = (
    pd.DataFrame(
        categorical_transformer.fit_transform(df_data_processing),
        columns = categorical_transformer.get_feature_names_out(),
        index = df_data_processing.index
    )
    .rename_columns(
        function = lambda x: x.removeprefix("onehotencoder__")
    )
    .rename_columns(
        function = lambda x: x.removeprefix("remainder__")
    )
)

df_data_processing
```

```python
df_data_processing.info()
```

ESCALABILIDAD DE DATOS

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_data_processing_scaled =  scaler.fit_transform(df_data_processing)
df_data_processing_scaled
```

```python
df_data_processing_scaled = pd.DataFrame(df_data_processing_scaled)
df_data_processing_scaled.columns = df_data_processing.columns
df_data_processing_scaled
```

# ANALISIS DE CORRELACIÓN

```python
sns.heatmap(
    data=df_data_processing_scaled.corr(),
    cmap=sns.diverging_palette(20, 230, as_cmap=True),
    center=0,
    vmin=-1,
    vmax=1,
    linewidths=0.5,
    cbar_kws={"shrink": 0.5},
    annot=True
)
```

```python
fig = plt.figure(figsize=(15,9))
df_data_processing_scaled.corr()['Churn'].sort_values(ascending=True).plot(kind='bar')
plt.show()
```

```python
sns.countplot(data=df_data,x='gender',hue='Churn')
plt.show()
```

```python
def plot_categorical(column):
  fig = plt.figure(figsize=(10,10))
  sns.countplot(data=df_data,x=column,hue='Churn')
  plt.show()

column_cat = df_data.select_dtypes(include='object').columns

for _ in column_cat:
  plot_categorical(_)
```

MODELO DE REGRESIÓN LOGISTICA

```python
df_data_processing_scaled.dropna(inplace=True)
```

```python
X = df_data_processing_scaled.drop('Churn',axis=1)
y = df_data_processing_scaled['Churn'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
result = model.fit(X_train, y_train)
```

```python
from sklearn import metrics
prediction_test = model.predict(X_test)
print(metrics.accuracy_score(y_test,prediction_test ))
```

# EVALUACION DEL MODELO

```python
model.predict_proba(X_test)
```

```python
model.coef_
```

```python
weights = pd.Series(model.coef_[0],
                    index=X.columns.values)
print(weights.sort_values(ascending=False)[:10].plot(kind='bar'))
```
