```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
sns.set_style(style='whitegrid')
sns.set_context(context='notebook')
plt.rcParams['figure.figsize'] = (11, 9.4)

penguin_color = {
    'Adelie': '#ff6602ff',
    'Gentoo': '#0f7175ff',
    'Chinstrap': '#c65dc9ff'
}

penguins_df = sns.load_dataset('penguins')
penguins_df
```

## RELACIÓN ENTRE 2 VARIABLES(GRAFICO DE PUNTOS)

```python
sns.scatterplot(
    data=penguins_df,
    x='bill_length_mm',
    y='bill_depth_mm',
    alpha=1/2,
    s=100
)
```

```python
sns.displot(
    data=penguins_df,
    x='bill_length_mm',
    y='bill_depth_mm'
)
```

```python
sns.displot(
    data=penguins_df,
    x='bill_length_mm',
    y='bill_depth_mm',
    kind='kde'
)
```

```python
sns.jointplot(
    data=penguins_df,
    x='bill_length_mm',
    y='bill_depth_mm',
    marginal_kws=dict(bins=25,fill=False)
)
```

RELACIÓN ENTRE VARIABLES NÚMERICAS Y CATEGORICAS

```python
sns.scatterplot(
    data=penguins_df,
    x='species',
    y='flipper_length_mm',
    hue='species',
    palette=penguin_color
)
```

GENERANDO RUIDO EN LOS VALORES

```python
sns.stripplot(
    data=penguins_df,
    x='species',
    y='flipper_length_mm',
    palette=penguin_color
)
```

BOXPLOT

```python
ax = sns.boxplot(
    data=penguins_df,
    x='flipper_length_mm',
    y='species',
    palette=penguin_color
)

ax = sns.stripplot(
    data=penguins_df,
    x='flipper_length_mm',
    y='species',
    color='.3'
)
```

GRAFICO DEL VIOLIN

```python
ax = sns.violinplot(
    data=penguins_df,
    x='species',
    y='flipper_length_mm',
    color='.8'
)

ax = sns.stripplot(
    data=penguins_df,
    x='species',
    y='flipper_length_mm',
    palette=penguin_color
)
```

SWARMPLOT

```python
sns.swarmplot(
    data=penguins_df,
    x='species',
    y='flipper_length_mm',
    hue='species',
    palette=penguin_color
)
```
