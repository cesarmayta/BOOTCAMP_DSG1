# EJEMPLO DE USO DEL KERNEL(FILTRO) EN IMAGENES

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage as nd
from skimage import io, color
```

# CARGAR IMAGEN

```python
img = io.imread('/content/foto.jpg')
plt.imshow(img)

```

# CONVERTIR IMAGEN

```python
img_gray = color.rgb2gray(img)
plt.imshow(img_gray)
```

# CREAMOS UN FILTRO(KERNEL)

```python
kernel = np.array([[-1,0,1],
                   [-1,0,1],
                   [-1,0,1]])
kernel
```

```python
kernel = np.array([[ 0, 0, 0],
                   [ 0, 1, 0],
                   [ 0, 0, 0]])
```

```python
kernel_blur = np.array([[ 0.0625, 0.125, 0.0625],
                   [ 0.125, 0.25, 0.125],
                   [ 0.0625, 0.125, 0.0625]])
```

# APLICAMOS CONVOLUCIÓN USANDO EL KERNEL

```python
img_new = nd.convolve(img_gray,kernel)
plt.imshow(img_new)
```
