```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
df_books = pd.read_csv('/content/drive/MyDrive/BOOTCAMP_DSG1/PANDAS/libros.csv')
df_books.head(2)
```

CREAR GRAFICO DE BARRAS PARA MOSTRAR LA SUMATORIA DE REVIEWS POR AÑO

```python
df_year_reviews_sum = df_books.groupby('Year')['Reviews'].sum()
df_year_reviews_sum
```

```python
import matplotlib.pyplot as plt
df_year_reviews_sum.plot(kind='bar',color=['red','green','yellow'])
plt.title('Total de Reviews por año')
plt.xlabel('Año')
plt.ylabel('Reviews')
plt.show()
```
