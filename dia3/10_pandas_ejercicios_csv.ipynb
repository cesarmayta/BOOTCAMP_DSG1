```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
df_books = pd.read_csv('/content/drive/MyDrive/BOOTCAMP_DSG1/PANDAS/libros.csv')
df_books.head(2)
```

CREAR GRAFICO DE BARRAS PARA MOSTRAR LA SUMATORIA DE REVIEWS POR AÑO

```python
df_year_reviews_sum = df_books.groupby('Year')['Reviews'].sum()
```

```python
import matplotlib.pyplot as plt
colors = ['red','green','blue']
df_year_reviews_sum.plot(kind='bar',color=colors)
plt.title('Total de Reviews por año')
plt.xlabel('Año')
plt.ylabel('Reviews')
plt.show()
```

MOSTRAR GRAFICO DEL RATING MAS ALTO POR AUTHOR

```python
df_author_rating_max = df_books.groupby('Author')['User Rating'].max()
df_author_rating_max
```

```python
import matplotlib.pyplot as plt
colors = ['red','green','blue']
df_author_rating_max.plot(kind='bar',color=colors)
plt.title('Rating mas alto por Author')
plt.xlabel('Author')
plt.ylabel('Rating')
plt.show()
```
