# IMPORTAMOS LIBRERIAS

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# CAMBIOS ESTILO DE GRAFICOS

```python
%matplotlib inline
sns.set_style(style='whitegrid')
sns.set_context(context='notebook')
plt.rcParams['figure.figsize'] = (11,9.4)

penguin_color ={
    'Adelie': '#ff6602ff',
    'Gentoo': '#0f7175ff',
    'Chinstrap':'#c65dc9ff'
}
```

# CARGAMOS DATASET

```python
penguins_df = sns.load_dataset('penguins')
penguins_df
```

# CONTEO DE VARIABLES

## MOSTRAR TODAS LAS VARIABLES

```python
penguins_df.describe(include='all')
```

## MOSTRAR VARIABLES NÚMERICAS

```python
penguins_df.describe(include=[np.number])
```

## MOSTRAR VARIABLES CATEGORICAS

```python
penguins_df.describe(include=object)
```

## CONVERTIR OBJECT A CATEGORY

```python
penguins_df.dtypes
```

```python
penguins_df = (
    penguins_df
    .astype({
        'species':'category',
        'island':'category',
        'sex':'category'
    })
)
```

```python
penguins_df.dtypes
```

```python
penguins_df.describe(include=['category'])
```

# ANALISIS DE VARIABLE SPECIES

## MOSTRAR CONTEOS DE VARIABLE CATEGORICA SPECIES

```python
(
    penguins_df
    .species
    .value_counts()
    .plot(
        kind='bar',
        color=penguin_color.values()
    )
)
```

## MOSTRAR PROPORCIONES DE VARIABLE SPECIES

```python
penguins_df['variable'] = ''
penguins_df
```

```python
(
    penguins_df
    .pipe(
        lambda df:(
            sns.displot(
                data=df,
                x='variable',
                hue='species',
                multiple='fill',
                palette=penguin_color
            )
        )
    )
)
```
