
**01 - IMPORTAMOS LIBRERIAS**



```python
import requests
import matplotlib.pyplot as plt
```

**02 - HACER SOLICITUD AL API PARA OBTENER DATOS**

```python
url = "https://randomuser.me/api/?results=10"
response = requests.get(url)

if response.status_code == 200:
  data = response.json()
else:
  print(f"ERROR : {response.status_code}")
```

**03 - EXTRAER EDADES DE LAS PERSONAS**

```python
edades = [persona['dob']['age'] for persona in data['results']]
edades
```

**04 - CREAR GRAFICO DE BARRAS PARA MOSTRAR EDADES**

```python
plt.bar(range(1,len(edades) + 1),edades,color='skyblue')
plt.title("Edades de personas generadas aleatoriamente")
plt.xlabel("Persona")
plt.ylabel('Edad')
plt.show()
```
