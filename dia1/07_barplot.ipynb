```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
paises = ['PERÚ','COLOMBIA','MEXICO','BOLIVIA','CHILE']
poblacion = [1000,800,900,1000,300]

plt.bar(paises,poblacion,width=0.2,color=['red','blue','green'])
plt.xticks(np.arange(5),('Perú','Colombia','Mexico','Bolivia','Chile'),rotation=45)
plt.show()
```

```python
plt.barh(paises,poblacion)
plt.show()
```
