```python
import seaborn as sns
import matplotlib.pyplot as plt
```

CARGAMOS UN DATASET DE EJEMPLO

```python
tips = sns.load_dataset('tips')
tips
```

CREAMOS UN DISPLOT

```python
sns.displot(data=tips,x='total_bill')
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',y='tip')
plt.show()
```

HUE PARA AGRUPAMIENTO

```python
sns.displot(data=tips,x='total_bill',hue='sex')
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',hue='sex',kind='kde',legend=True,palette='dark')
plt.show()
```

LEYENDAS

```python
sns.displot(data=tips,x='total_bill',hue='smoker',kind='kde',legend=True,palette='dark')
plt.title('Total de propinas por fumadores y no fumadores')
plt.xlabel('Total de propina')
plt.ylabel('Densidad')
plt.show()
```
