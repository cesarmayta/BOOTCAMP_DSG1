```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
tips.head()
```

HISTPLOT

```python
sns.histplot(data=tips,x='tip',hue='sex',cumulative=False,stat='count',multiple='dodge')
plt.show()
```

KDEPLOT

```python
sns.kdeplot(data=tips,x='tip',hue='sex',shade=True,bw_adjust=1)
plt.show()
```
