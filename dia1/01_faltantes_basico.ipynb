```python
import pandas as pd
```

CREAMOS UN DATAFRAME DE PRUEBA

```python
df = pd.DataFrame.from_dict(
    data = {
        "a": list('asdfasdfas'),
        "b": range(0,10)
    }
)
df
```

GENERAMOS NULOS

```python
df.iloc[2:5,0] = None
df.iloc[6:7,1] = None
df
```

VERIFICAMOS NULOS

```python
df.isna()
```

CONTEO DE NULOS

```python
df.isna().sum()
```

```python
df.isna().sum().sum()
```

DIFERENCIA ENTRE TOTAL Y NULOS

```python
df.size - df.isna().sum().sum()
```

CONTEO POR CASOS

```python
df.isna().pipe(
    lambda df_1 : (
        df_1.sum()
        .reset_index(name='n_missing')
        .rename(columns={'index':'variable'})
        .assign(
            n_cases=len(df_1),
            pct_missing=lambda df_2: df_2.n_missing / df_2.n_cases * 100,
        )
    )
)
```

EXTENDIENDO API DE PANDAS

```python
@pd.api.extensions.register_dataframe_accessor("missing")
class MissingMethods:

    def __init__(self,pandas_obj):
      self._df = pandas_obj

    def number_missing(self):
      return self._df.isna().sum().sum()
```

```python
df = pd.DataFrame.from_dict(
    data = {
        "a": list('asdfasdfas'),
        "b": range(0,10)
    }
)
df.iloc[2:5,0] = None
df.iloc[6:7,1] = None
```

```python
df.missing.number_missing()
```
